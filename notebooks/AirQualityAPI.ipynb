```python
import requests

# URL til MET API for luftkvalitet
API_URL = "https://api.met.no/weatherapi/airqualityforecast/0.1/"

# Header med "User-Agent" (påkrevd av MET API)
HEADERS = {
    "User-Agent": "Georg Rommetveit/0.1 (georg.romm@gmail.com)"  # Erstatt med dine detaljer
}

# Sett koordinater
params = {
    "lat": 59.9139,  # Breddegrad (latitude)
    "lon": 10.7522,  # Lengdegrad (longitude)
}

# Send forespørselen til API-et
response = requests.get(API_URL, headers=HEADERS, params=params)

# Sjekk om forespørselen var vellykket
if response.status_code == 200:
    data = response.json()
    print("Luftkvalitetsdata:", data)
else:
    print(f"Feil: {response.status_code}, {response.text}")
```
